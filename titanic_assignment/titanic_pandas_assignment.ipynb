```python
def value_count_with_pct(value):
    count=value.value_counts()
    percent=value.value_counts(normalize=True)*100
    return pd.DataFrame({'count':count,'percentage':percent})
```

```python
passanger_class=value_count_with_pct(df['Pclass'])
passanger_class
```

```python
gender = value_count_with_pct(df['Sex'])
gender
```

```python
embark = value_count_with_pct(df['Embarked'])
embark
```

```python
female_firstclass_over_30=df[(df['Pclass']==1) & (df['Sex']=='female') & (df['Age']>=30)]
female_firstclass_over_30.sort_values(by='Fare',ascending=False).head(20)
```

```python
age_mean = df['Age'].mean()
print(age_mean)
age_median = df['Age'].median()
print(age_median)
age_mode = df['Age'].mode()
print(age_mode)
```

```python
df.groupby('Pclass')['Fare'].mean()
```

```python
df.groupby('Sex')['Survived'].mean()*100
```

```python
df['Age_imputed'] = df['Age']
df['Age_imputed'] = df.groupby(['Pclass','Sex'])['Age_imputed'].transform(lambda x: x.fillna(x.median()))
df
```

```python
df['Title'] = df['Name'].str.extract('([A-Za-z]+)\.', expand=False)
```

```python
df['Title'].value_counts()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = df['FamilySize']==1
df
```
